## Create a scatter plot

### Import matplotlib library

```python
import matplotlib.pyplot as plt
```

### Access column data separately

```python
life_exp = df_gapminder.life_expectancy
money_pp = df_gapminder.money_per_person
countries = df_gapminder.country
```

### Pass data to scatter function

```python
plt.scatter(x=life_exp, y=money_pp)
```

## Customize the plot

### Change the size

```python
plt.scatter(x=life_exp, y=money_pp, s=20)
```

### Change the color

```python
plt.scatter(x=life_exp, y=money_pp, s=20, c="red")
```

### Change the marker

```python
plt.scatter(x=life_exp, y=money_pp, s=20, c="red", marker="X")
```

### Change the border

```python
plt.scatter(x=life_exp, y=money_pp, s=200, c="red", marker="X", edgecolor="green")
```

```python
px.scatter(data_frame=df_gapminder, x="life_expectancy", y="money_per_person", hover_name="country", color="continent")
```
